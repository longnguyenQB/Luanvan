```python
col = ['phone', 'phone_member', 'type','time', 'call_to', 'call_in', 'in_contact', 'duration', 'same_network', 'in_hour', 'success']
df1 = df_root[col]
```

# phone_member -> phone_to by type :

```python
df1[['phone',
     'phone_member']] = df1[['phone_member',
                           'phone']].where(df1['type'].isin([2, 4]),
                                           df1[['phone', 'phone_member']].values)
df1[['call_to',
     'call_in']] = df1[['call_in',
                           'call_to']].where(df1['type'].isin([2, 4]),
                                           df1[['call_to', 'call_in']].values)
```

```python
df1 = df1.rename(columns={"phone" : "phone_to", "phone_member":"phone_in"})
df1 = df1.drop(labels=['type'], axis=1)
df1 = df1.sort_values(by=['time'])
```

```python
# def convert_time(df_time):
# 	df_time = datetime.fromisoformat(df_time)
# 	return df_time
# df1['time']=df1.apply(lambda x:convert_time(x['time']),axis=1)
```

```python
def frequency_call(df_time):
    freq = np.median(np.diff(df_time))
    return (freq / np.timedelta64(1, 's')) / 3600
```

# df2 for edge

```python
df2 = df1.groupby(by=['phone_to', 'phone_in']).agg({
    'in_contact':
    'mean',
    'same_network':
    'mean',
    'in_hour':
    'sum',
    'success':
    'sum',
    'duration':
    'sum',
    'call_to':
    'sum',
    # 'call_in':
    # 'sum',
    'time':
    lambda x: frequency_call(x)
})
df2 = df2.reset_index()
df2.time = df2.time.fillna(0)
```

# df3 for node

```python
df3 = df1.groupby(by=['phone_to']).agg({
    'in_contact':
    'sum',
    'same_network':
    'sum',
    'in_hour':
    'sum',
    'success':
    'sum',
    'duration':
    'sum',
    'call_to':
    'sum',
    'call_in':
    'sum',
    'time':
    lambda x: frequency_call(x)
})
df3.time = df3.time.fillna(0)
```
